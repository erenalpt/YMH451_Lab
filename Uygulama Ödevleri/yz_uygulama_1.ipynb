İki tane (3,3) boyutunda rastgele sayılardan meydana bir dizi oluşturunuz. Oluşturulan bu diziyi hem yatay hem de dikey olacak şekilde istif (stack) ediniz 

```
arr_1 = np.random.randint(15, size=(3,3))  
arr_2 = np.random.randint(25, size=(3,3))  
print("Array-1",arr_1, sep='\n')
print("Array-2",arr_2,sep='\n')
## Horitzontal Stack
h_stack = np.hstack((arr_1,arr_2))

## Vertical Stack
v_stack = np.vstack((arr_1,arr_2))

print("Horizontal",h_stack, sep='\n')
print("Vertical",v_stack, sep='\n')
```
