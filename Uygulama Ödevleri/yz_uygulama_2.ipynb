```
print("Uygulama 2 - Soru 1")
try:
    aci_1 = int(input("1. Aciyi giriniz: "))
    aci_2 = int(input("2. Aciyi giriniz: "))
    aci_3 = int(input("3. Aciyi giriniz: "))
except:
    print("Lütfen girdiginiz degerleri kontrol edin.")    
if (aci_1+aci_2+aci_3)==180:
    if (aci_1<90 and aci_2<90 and aci_3<90):
        print("Dar Acili Ucgen")
    elif (aci_1== 90 or aci_2== 90 or aci_3== 90):
        print("Dik Ucgen")
    elif (aci_1 > 90 or aci_2 > 90 or aci_3 > 90):
        print("Genis Acili Ucgen")
else:
    print("Boyle bir Ucgen olamaz!")
```

```
print("Uygulama 2 - Soru 4")

try:
    yas = int(input("Yasinizi giriniz: "))
except:
    print("Lutfen tam sayi turunde bir deger giriniz!")
if (yas > -1 and yas<162):
    if (yas<2):
        print("Bu kisi bebektir.")
    elif (yas>1 and yas<4):
        print("Bu kisi yeni yürümeye baslayan cocuktur.")
    elif (yas>3 and yas <13):
        print("Bu kisi cocuktur.")
    elif (yas>12 and yas<20):
        print("Bu kisi ergendir.")
    elif (yas>19 and yas<65):
        print("Bu kisi yetiskindir.")
    elif (yas>64 and yas<162):
        print("Bu kisi yaslidir.\n 160'a kadar yasamissa eger Zaro Ağayi da gecmistir.")
else:
    print("Zaro Ağa dan da yasli insan yoktur heralde =) ")
```
