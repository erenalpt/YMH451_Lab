```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/erenalpt/YMH451_Lab/main/Mobile_Price.csv")
```

```python
df.head()
```

```python
label_encoder = LabelEncoder().fit(df.price_range)
labels=label_encoder.transform(df.price_range)
classes = list(label_encoder.classes_)
```

```python
X=df.drop(["price_range", "blue", "fc", "int_memory", "ram", "wifi"], axis=1)
y = labels
```

```python
X.head()
```

```python
print(X.shape)
print(y.shape)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(16, input_dim=X_train.shape[1], activation="relu"))
model.add(Dense(32, activation="relu"))
model.add(Dense(32, activation="relu"))
model.add(Dense(4, activation="softmax"))
model.summary()

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100)
```

```python
print("Mean Train Loss: ", np.mean(model.history.history["loss"]))
print("Mean Train Accuracy: ", np.mean(model.history.history["accuracy"]))
print("Mean Test Loss: ", np.mean(model.history.history["val_loss"]))
print("Mean Test Accuracy: ", np.mean(model.history.history["val_accuracy"]))
```

```python
plt.plot(model.history.history["accuracy"])
plt.plot(model.history.history["val_accuracy"])
plt.title("Modal Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epok")
plt.legend(["Train", "Test"], loc="upper left")
plt.show
```

```python
plt.plot(model.history.history["loss"])
plt.plot(model.history.history["val_loss"])
plt.title("Modal Loss")
plt.ylabel("Loss")
plt.xlabel("Epok")
plt.legend(["Train", "Test"], loc="upper left")
plt.show
```

```python
kfold = KFold(n_splits=3, shuffle=True, random_state=0)
scores=cross_val_score(model, X, y, cv=kfold)
print(scores) 
print(scores.mean()) 
```
